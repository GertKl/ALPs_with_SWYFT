```python
torch.multiprocessing.set_start_method('spawn', force=True)
```

```python
n = 1e8
```

```python
%%time
# n = 1e4
n_procs = 100
with Pool(n_procs) as pool: 
    pool.map(sleeper, [int(n/n_procs)]*n_procs)
```

```python
%%time
# n = 1e4
n_procs = 4
with Pool(n_procs) as pool: 
    pool.map(sleeper, [int(n/n_procs)]*n_procs)
```

```python
%%time
# n = 1e+4
n_procs = 4
with Pool(n_procs) as pool: 
    pool.map(sleeper2, [1]*int(n), chunksize=int(n/n_procs))
```
